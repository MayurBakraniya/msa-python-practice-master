```python
#EDA(Exploratory Data Analysis) of NewYork City 2019

#Libraries for Data Analysis & Wrangling

#pandas is a fast, powerful, flexible and easy to use open source data analysis and manipulation tool, built on top of the Python programming language.
import pandas as pd 

#NumPy is a library for the Python programming language, adding support for large, multi-dimensional arrays and matrices, along with a large collection of high-level mathematical functions to operate on these arrays.
import numpy as np

#Matplotlib is a comprehensive library for creating static, animated, and interactive visualizations in Python.
import matplotlib.pyplot as plt
%matplotlib inline #%matplotlib inline sets the backend of matplotlib to the 'inline' backend

#Seaborn is a library for making statistical graphics in Python
import seaborn as sns

#The OS module in python provides functions for interacting with the operating system.
import os

#importing csv file using pandas and read_csv command. 
nyc_data = pd.read_csv('E:\AB_NYC_2019.csv')


#head() method is used to print top 5 rows & tail() method is used to print last 5 row
nyc_data.head(10).append(nyc_data.tail())
```

```python
#printing the dataset
print(nyc_data)
```

```python
#printing the information of the NYC dataset using info() method
print(nyc_data.info())
```

```python
#getting dataypes using dtypes() method
print(nyc_data.dtypes)
```

```python
#statistics of the given dataest
# we use describe() method

print(nyc_data.describe())
```

```python
#finding null values using isnull() method
print(nyc_data.isnull())
```

```python
#calculating the total null values dataset
print(nyc_data.isnull().sum())

#we've seen that there're 16 null values in names, 21 null values in host_name, 10052 null values in last_review & reviews_per_month
```

```python
#Data cleaning

missing_values = nyc_data.isnull().sum().sort_values(ascending=False)

percentage_missing_values = round((nyc_data.isnull().sum()/nyc_data.isnull().sum().sum()).sort_values(ascending=False),3)

null_in_data = pd.concat([missing_values, percentage_missing_values],keys=['Missing values','Percentage'],axis=1)
null_in_data.head()

#as we see that both reviews_per_month & last_review have same missing values.
```

```python
#bar plot for missing values(reviews_per_month, last_review) using barplot() method

plt.figure(figsize=(10,5))
missing_plot = sns.barplot(x=missing_values.index[0:4],y=missing_values[0:4],palette='bright')
missing_plot.set_xticklabels(missing_plot.get_xticklabels(),rotation=0)
plt.title('Missing values in dataset.')
plt.show()
```

```python
#As per my knowledge it's best to drop columns that are not significant or could be unethical to use for our future data exploration and predictions

nyc_data.drop(['id','host_name','last_review'], axis=1, inplace=True)
nyc_data
```

```python
#replacing all NaN values in 'reviews_per_month' with 0
nyc_data.fillna({'reviews_per_month':0}, inplace=True)

nyc_data.reviews_per_month.isnull().sum()
```

```python
print(nyc_data['reviews_per_month'].tail(20))
```

```python
#let's proceed with examing some interesting categorical unique values
#examining the unique values of neighbourhood_group as this column will appear very handy for later analysis
#unique() function is used to get unique values of Series object. 

nyc_data.neighbourhood_group.unique()
```

```python
#printing pie chart using pie() method

labels = nyc_data.neighbourhood_group.value_counts().index
colors = ['lightblue','beige','lightgreen','orange','cyan']
explode = [0,0,0,0,0]
sizes = nyc_data.neighbourhood_group.value_counts().values

plt.figure(0,figsize = (7,7))
plt.pie(sizes, explode=[0.1,0.0,0.3,0.5,0.0], labels=labels, colors=colors, autopct='%1.1f%%',shadow=True)
plt.title('Neighbourhood Group',color = 'black',fontsize = 15)
plt.show()
```

```python
#neighbourhood_group-price
#printing barplot of price and neighbourhood_group
result = nyc_data.groupby(["neighbourhood_group"])['price'].aggregate(np.median).reset_index().sort_values('price')
sns.barplot(x='neighbourhood_group', y="price", data=nyc_data,palette=colors, order=result['neighbourhood_group']) 
plt.xticks(rotation=45)
plt.show()

#As we see that Manhattan has highest price
```

```python
#neighbourhood_group-availability_365
#printing the barplot between neighbourhood_group whose availability is 365 

result = nyc_data.groupby(["neighbourhood_group"])['availability_365'].aggregate(np.median).reset_index().sort_values('availability_365')
sns.boxplot(x='neighbourhood_group', y="availability_365", data=nyc_data) 
plt.show()
```

```python
#Piechart for rooms

labels = nyc_data.room_type.value_counts().index
colors = ['lightblue','pink','beige']
explode = [0,0,0]
sizes = nyc_data.room_type.value_counts().values


plt.figure(0,figsize = (7,7))
plt.pie(sizes, explode=[0,0.05,0.5], labels=labels, colors=colors, autopct='%1.1f%%', shadow=True)
# plot.pie(explode=,autopct='%1.1f%%',ax=ax[0],)
plt.title('Room-Type',color = 'Brown',fontsize = 15)
plt.show()
```

```python
#room_type-price
result = nyc_data.groupby(["room_type"])['price'].aggregate(np.median).reset_index().sort_values('price')
sns.barplot(x='room_type', y="price", data=nyc_data, order=result['room_type']) 
plt.show()
```

```python
#room_type-availability_365

result = nyc_data.groupby(["room_type"])['availability_365'].aggregate(np.median).reset_index().sort_values('availability_365')
sns.boxplot(x='room_type', y="availability_365", data=nyc_data, order=result['room_type']) 
plt.show()
```

```python
#countplot between rooms in different neighbourhood_group

plt.figure(figsize=(14,8))
sns.countplot(nyc_data.sort_values('neighbourhood_group').neighbourhood_group,hue=nyc_data.room_type)
plt.title('Borough wise room type count')
plt.xlabel('Borough name')
plt.ylabel('Room Counts')
plt.show()
```

```python
# Lineplot between availability_365 & price using lineplot() method

sns.lineplot(x='availability_365',y='price',data=nyc_data)
plt.show()
```

```python
#printing scatterplot using scatterplot() method

plt.figure(figsize=(10,6))
sns.scatterplot(nyc_data.longitude,nyc_data.latitude,hue=nyc_data.neighbourhood_group)
plt.ioff() #The ioff() function in pyplot module of matplotlib library is used to turn the interactive mode off.
```
